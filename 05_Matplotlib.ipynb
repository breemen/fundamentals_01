# Matplotlib

There are many packages to visualize data in Python. One of the most used package is [Matplotlib](https://matplotlib.org/). This package can be used to visualize data from Python, Numpy and Pandas.

```python
# import numpy, pandas en matplotlib
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
mpl.__version__
```

```python
# basic plot
x = np.linspace(-1, 1, 10)
y = x**2
plt.plot(x, y);
```

```python
# Matplotlib provides different plotting styles
plt.style.available
```

```python
# classical style
with plt.style.context(('classic')):
    plt.plot(x, y);
```

```python
# modern stylee
with plt.style.context(('seaborn-whitegrid')):
    plt.plot(x, y);
```

```python
# plot examples in all available styles
for i in range(len(plt.style.available)):
    plt.figure(figsize=(4, 3))
    with plt.style.context((plt.style.available[i])):
        plt.plot(x, y)
        plt.title(plt.style.available[i])
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
```

## Matlab interface vs object-oriented interface

Matplotlib provides two interfaces to create graphs. The first method is based on the 'stateful' plot functionality of Matlab. Below you see an example.

```python
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x));
```

The second method is object oriented, as illustrated below.

```python
fig, ax = plt.subplots(2, figsize=(8,6))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

Here, the `ax` object is used to plot. In the 'stateful Matlab' interface the state (o.a. where to plot) is remembered internally by the matplotlib package.

## Graph design

Matplotlib provides many options to plot line graphs.

```python
fig, ax = plt.subplots(3, figsize=(8,8))
# line color: 'blue', '0.6' (grayscale), '(0.2, 0.4, 0.5)' (rgb). '#AA02FF' Hex code RRGGBB
ax[0].plot(x, np.sin(x), color='blue')
# line style: '-', '--', '-.', ':'
ax[1].plot(x, np.cos(x), linestyle=':')
# marker style: 'o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd'
ax[2].plot(x, np.cos(x), marker='x');
```

```python
# Labels and limits
fig, ax = plt.subplots(2, figsize=(8,6))
ax[0].plot(x, np.sin(x), color='blue')
ax[0].set_xlim(-2, 12)
ax[0].set_ylim(-1.2, 1.2)
ax[0].set_title("sin")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[1].plot(x, np.cos(x), linestyle=':')
ax[1].set_xlim(-2, 12)
ax[1].set_ylim(-1.2, 1.2)
ax[1].set_title("cos")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
# voorkom overlap van labels met tight_layout()
plt.tight_layout()
```

```python
# Complete graph example
fig, ax = plt.figure(figsize=(8,4)), plt.axes()
ax.plot(x, np.sin(x), color='blue', label="sin(x)")
ax.plot(x, np.cos(x), linestyle=':', label="cos(x)")
ax.set(xlim=(-2, 12), ylim=(-1.2, 1.2), title="Example line graph", xlabel="x", ylabel="y")
plt.legend();
```

The grid of a plot can be turned off using `plt.grid()`.

```python
fig, ax = plt.subplots(2, 1, figsize=(8,6))

ax[0].plot(x, np.sin(x), color='blue')
ax[0].grid(False)
ax[0].set(xlim=(-2, 12), ylim=(-1.2, 1.2), title="No grid", xlabel="x", ylabel="y")
ax[1].plot(x, np.sin(x), color='blue')
ax[1].grid(True, linewidth=3, color='r', linestyle='-.')
ax[1].set(xlim=(-2, 12), ylim=(-1.2, 1.2), title="Custom grid style", xlabel="x", ylabel="y")

# adjust layout to avoid overlap
plt.tight_layout();
```

## Scatter plot

```python
x = np.random.normal(0, 2, size=100)
y = np.random.normal(0, 1, size=100)
fig, ax = plt.figure(figsize=(5, 5)), plt.axes()
ax.scatter(x, y)
ax.set(xlim=(-6,6), ylim=(-6,6), xlabel="x", ylabel="y", title="scatter plot");
```

The marker type of a scatter plot can be adjusted. For a complete overview see https://matplotlib.org/api/markers_api.html.

```python
from matplotlib.markers import MarkerStyle
MarkerStyle.markers
fig, ax = plt.subplots(figsize=(12, 12))
ncols = 6
for i, m in enumerate(MarkerStyle.markers.keys()):
    ax.scatter(i%ncols, -int(i/ncols), marker=m, s=100)
    ax.text(i%ncols, -int(i/ncols)-0.4, '{} : {}'.format(m,MarkerStyle.markers[m]), ha="center")
ax.set(xlim=(-0.15, ncols + 0.15), ylim=(-int(len(MarkerStyle.markers)/ncols)-0.15, 0.15), title="type markers")
plt.axis('off');
```

```python
# create an array with a color for every data point
colors = 1 - (np.abs(x) / np.max(np.abs(x)))

# create an array with a marker size for every data point
sizes  = 500 * (np.abs(y) / np.max(np.abs(y)))

# create figure
fig, ax = plt.figure(figsize=(5, 5)), plt.axes()

# create plot
# select colormap (cmap), see https://matplotlib.org/gallery/color/colormap_reference.html
# or https://matplotlib.org/examples/color/colormaps_reference.html
im = ax.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='inferno')
ax.set(xlim=(-6,6), ylim=(-6,6), xlabel="x", ylabel="y", title="scatter plot")

# add colorbar
fig.colorbar(im, ax=ax, orientation='horizontal');
```

## Subplots with different sizes

To plot multiple figure with the same size use `plt.subplots()`. To plot graphs with different sizes, use the [GridSpec](https://matplotlib.org/api/_as_gen/matplotlib.gridspec.GridSpec.html#matplotlib.gridspec.GridSpec) object.

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(2, 2, height_ratios=[2, 1])
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1, 0])
ax3 = plt.subplot(gs[1, 1])

x = np.linspace(0, 10, 25)
ax1.plot(x, np.sin(2 * np.pi * x / 5), '.');
ax2.plot(x, np.sin(2 * np.pi * x / 10), '-');
ax3.plot(x, np.cos(2 * np.pi * x / 10), '--');
```

# 3D plots

Matplotlib also support plotting 3D data.

## Plot surface

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2 - 0.5*x*y

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, rstride=2, cstride=2, cmap='viridis', linewidth=0)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# define elevation (z) en azimuth (x,y) of Axes object
ax.view_init(40, 60)
```

## Plot 3d line

```python
z = np.linspace(0, 10, 100)
x = 10 * np.sin(z)
y = 10 * np.cos(z)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
p = ax.plot(x, y, z, c='r', lw=2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# define elevation (z) en azimuth (x,y) of Axes object
ax.view_init(40, 60)
```

## Plot contour plot

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2 - 0.5*x*y

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.contour(x, y, z, 15, linewidths = 0.5, colors = 'k')
p = ax.pcolormesh(x, y, z, cmap = plt.get_cmap('viridis'))
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(p, ax=ax);
```
